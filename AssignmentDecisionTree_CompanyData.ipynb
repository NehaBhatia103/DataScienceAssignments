#Preprocessing

```python
data["Urban"]=data["Urban"].replace(["Yes","No"],[1,0])
data["US"]=data["US"].replace(["Yes","No"],[1,0])
data["ShelveLoc"]=data["ShelveLoc"].replace(["Good","Medium","Bad"],[1,2,3])
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
scaled_data
```

```python
x=scaled_data.iloc[:,1:9]
y=scaled_data["Sales"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=3)
model.fit(x_train,y_train)
```

```python
from sklearn import tree
tree.plot_tree(model)
```

```python
data.columns
```

```python
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price','ShelveLoc', 'Age', 'Education']
cn=["Sales"]
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(5,5),dpi=100)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```

Predictions on test Data

```python
preds=model.predict(x_test)
```
